```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR, SVC
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.svm import SVR, SVC
from sklearn.metrics import accuracy_score, precision_score, classification_report, confusion_matrix, ConfusionMatrixDisplay, roc_curve, roc_auc_score
```

```python
df = pd.read_csv("nigeria_houses.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
categorical_cols = df.select_dtypes(include=["object"]).columns.tolist()

categorical_cols
```

```python
X = df.drop(columns=['price'])
y_reg = df['price']
```

```python
X_lab = X.copy()
```

```python
for col in categorical_cols:
    le = LabelEncoder()
    X_lab[col] = le.fit_transform(X_lab[col])
```

```python
X_lab
```

```python
scaler = StandardScaler()

scaledX = scaler.fit_transform(X_lab)
```

```python
scaledX_train, scaledX_test, y_reg_train, y_reg_test = train_test_split(scaledX, y_reg, test_size=0.2, random_state=42) #80:20 split
```

```python
scaledX_train.shape
```

```python
scaledX_test.shape
```

```python
print(scaledX_test.shape)
print(y_reg_test.shape)
```

```python
linear_model = LinearRegression()
linear_model.fit(scaledX_train, y_reg_train)
```

```python
linear_predictions = linear_model.predict(scaledX_test)
linear_predictions
```

```python
forest_reg = RandomForestRegressor(random_state=16)
forest_reg.fit(scaledX_train, y_reg_train)
```

```python
r2_score(y_reg_test, linear_predictions)
```

```python
forest_preds = forest_reg.predict(scaledX_test)
forest_preds
```

```python
r2_score(y_reg_test,forest_preds)
```

```python
# Define evaluation metrics
metrics = {
    "Mean Absolute Error": mean_absolute_error,
    "Mean Squared Error": mean_squared_error
}

results = {}
for name, func in metrics.items():
    # Calculate scores for each model
    linear_score = func(y_reg_test, linear_predictions)
    forest_reg_score = func(y_reg_test, forest_preds)

    # Store results in the dictionary
    results[name] = {
        "Linear Regression": linear_score,
        "Random Forest_reg": forest_reg_score
    }
```

```python
results
```

## The MSE for Random forest regression is less tha that of Linear regresssion hence Random forest regression is better for this mode

# CLASSIFICATION

```python
X = df.drop(columns=["price Category"])
y = df["price Category"]
```

```python
X
```

```python
X_lab = X.copy()
```

```python
for col in categorical_cols:
    le = LabelEncoder()
    X_lab[col] = le.fit_transform(X_lab[col])
```

```python
X_lab
```

```python
scaler = StandardScaler()

ScaledX = scaler.fit_transform(X_lab)
```

```python
ScaledX
```

```python
ScaledX_train, ScaledX_test, y_train, y_test = train_test_split(ScaledX, y, test_size=0.2, random_state=42)
```

```python
print(ScaledX_test.shape)
print(y_test.shape)
```

```python
log = LogisticRegression()
log.fit(ScaledX_train, y_train)
```

```python
log_preds = log.predict(ScaledX_test)
```

```python
metrics = {
    "Accuracy": accuracy_score,
    "Precision": precision_score
}

results = {}
for name, func in metrics.items():
    logistic_score = func(y_test, log_preds)
    results[name] = {"Logistic Regression": logistic_score}
```

```python
results
```

```python
svm = SVC()
svm.fit(ScaledX_train, y_train)
```

```python
svm_preds = svm.predict(ScaledX_test)
```

```python
svm_preds
```

```python
metrics = {
    "Accuracy": accuracy_score,
    "Precision": precision_score
}

results = {}
for name, func in metrics.items():
    svm_score = func(y_test, svm_preds)
    results[name] = {"SVC": svm_score}
```

```python
results
```

```python
print("Model Evaluation Results:\n")
for name, func in metrics.items():
    logistic_score = func(y_test, log_preds)
    svm_score = func(y_test, svm_preds)
    results[name] = {"Logistic Regression": logistic_score, "Support Vector Machine": svm_score}
    print(f"{name}:\n  Logistic Regression: {logistic_score:.3f}\n  Support Vector Machine: {svm_score:.3f}\n")

# Print summary of best model per metric
print("Best Model per Metric:")
for metric, scores in results.items():
    best_model = max(scores, key=scores.get)
    print(f"{metric}: {best_model}")
```

# Logistic regression is the best model for classification
